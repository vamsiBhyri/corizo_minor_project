```python
import pandas as pd
df = pd.read_csv("Stock_Price.csv")
print(df.columns)
df.isnull().sum()
```

Hence, we can conclude that there are no missing values/null values in the dataset.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
y = df.Close
features = ['Open','High','Low','Adj Close','Volume']
X = df[features]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X,y,random_state = 1)
model = DecisionTreeRegressor(max_leaf_nodes = 100, random_state = 1)
model.fit(train_X,train_y)
```

```python
predictions  = model.predict(val_X)
mae = mean_absolute_error(predictions, val_y)
print(mae)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(val_X.Open[:20],val_y[:20],color = 'green',label = 'Actual Values')
plt.scatter(val_X.Open[:20],predictions[:20],color = 'black',label = 'Predictions')
plt.legend()
plt.show()
```

As one can see, the closing values of the stock price are **accurately predicted** by the **Decision Tree Model** that is used.

## Clustering 

```python
#SCALING AND FITTING THE DATA AT HAND TO YIELD ACCURATE CLUSTERS
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['Open']])
df['Open'] = scaler.transform(df[['Open']])
scaler.fit(df[['Close']])
df['Close'] = scaler.transform(df[['Close']])
```

```python
#constructing the elbow plot to find the optimal amount of clusters
sse = []
k_range  = range(1,10)
for k in k_range:
    km = KMeans(n_clusters = k)
    km.fit(df[['Open','Close']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_range,sse)
plt.xlabel('K')
plt.ylabel('SSE')
```

From the above elbow plot, we can conclude that the number of clusters that is optimal as an input is **two**.

```python
#using the k means algorithm to cinstruct the clusters
km = KMeans(n_clusters = 2)
y_predicted = km.fit_predict(df[['Open','Close']])
y_predicted
```

```python
df['Cluster'] = y_predicted
df.columns
```

```python
df1 = df[df.Cluster == 0]
df2 = df[df.Cluster == 1]
plt.scatter(df1['Open'],df1['Close'],color = 'green',label = 'c1')
plt.scatter(df2['Open'],df2['Close'],color = 'red', label = 'c2')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker = '*',label = 'centriods')
plt.legend()
```

Here, we can clearly observe the **two clusters** present in our dataset.
